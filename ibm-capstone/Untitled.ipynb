## Dealing with Missing Values and Selecting Relevant Features

```python
#func to return the missing values for the features
def missingDataPerFeature(data) :
    missing_data = data.isnull().sum()
    missing_data_percentage = missing_data.apply(lambda x : x/data.shape[0]*100)
    return missing_data_percentage
```

```python
data_ = missingDataPerFeature(data)
data_
```

```python
#dropping features that have more than 50% of data mising
cols_to_drop = data_[data_>50].index.to_list()
cols_to_drop
```

```python
data = data.drop(columns=cols_to_drop, axis=1)
```
